# ch06_whileSum

```python
i, hap = 0, 0

i = 1

while i <= 10 :
    hap = hap + i
    i = i + 1
    
print("1에서 10까지의 합계 : %d " % hap)
```

# TwoSum

```python
hap = 0
num1, num2 = 0, 0

while True : 
    num1 = int(input("더할 첫 번째 정수 : "))
    if num1 == 0:
        break
    num2 = int(input("더할 두 번째 정수 : "))

    hap = num1 + num2

    print("%d + %d = %d" % (num1, num2, hap))

print("0을 입력해서 반복문을 탈출했습니다.")
```

# SumOver

```python
hap, i = 0, 0


for i in range(1, 101) :
    hap = hap + i
    if hap > 1000 :
        break
        
print("1~100의 합계를 최초로 1000이 넘을 때 더해진 수는 : %d" % i)
print("합계 : %d" % hap)

```

# ExceptThree

```python
hap, i = 0, 0

for i in range(1, 101, 1) :
    if i % 3 == 0 :
        continue
    hap = hap + i
    
print("1~100의 합계 : %d" % hap)
```

```python
hap, i = 0, 0

for i in range(1, 101, 1) :
    if i % 3 != 0 :
        hap = hap + i
    
print("1~100의 합계 : %d" % hap)
```

# DiamondStar

```python
i, k = 0, 0

while i < 9 :
    if i < 5 :
        
        k = 0
        while k < 4-i :
            print('  ', end="")
            k=k+1

        k = 0
        while k < 1+2*i :
            print('\u2605', end="")
            k = k + 1
    else :
        k = 0
        while k < i-4 :
            print('  ', end="")
            k=k+1

        k = 0
        while k < 17-2*i :
            print('\u2605', end="")
            k = k + 1
    print()
    i = i + 1
```

```python
for i in range(1, 6, 1) :
    for k in range(1, 6-i, 1) :
        print('  ', end="")
        
    for k in range(1, 2*i, 1) :
        print('\u2605', end="")
    print()
    
for j in range(0, 4, 1) :
    for k in range(1, 2+j, 1) :
        print('  ', end="")
        
    for k in range(1, 8-2*j, 1) :
        print('\u2605', end="")
    print()
```
